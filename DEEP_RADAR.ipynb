We have generated a set of signals with 23 modulations commonly used in radar applications. Each dataset item consists on a 1024x2 vector that represents the In-phase and Quadrature (I + jQ) time samples of a modulated signal with a noise contribuition. For each of the modulations, we considered Signal-to-Noise Ratio (SNR) values within the range (-12,20)dB with a step of 2 dB.
The following modulations have been considered as Classes:
# 1- Linear Frequency Modulation (LFM)
2- Binary Frequency-Shift Keying 2FSK
3- Frequency-Shift Keying 4FSK
4- Frequency-Shift Keying 8FSK
# 5- Frequency Modulation with Costas code
6- Binary Phase-Shift Keying 2PSK
7- Phase-Shift Keying4PSK
8- Phase-Shift Keying 8PSK
# 9- Phase Modulation with Barker code
10- Phase modulation with Huffman code
# 11- Phase modulation with Frank code
# 12- Phase modulation with P1 code
# 13- Phase modulation with P2 code
# 14- Phase Modulation with P3 code
# 15- Phase modulation with P4 code
16- Phase modulation with Px code
17- Phase modulation with Zadoff-Chu code
18- Phase modulation with T1 code
19- Phase modulation with T2 code
20- Phase modulation with T3 code
21- Phase modulation with T4 code
# 22- Non-modulation (NM)
# 23- Complex White Gaussian Noise (Noise)

```python
target_idx = [0, 4, 8, 10, 11, 12, 13, 14, 21, 22] # indices do filtro
```

```python
def filter_split(X, y, lbl, target_idx): #função do filtro
    mask = np.isin(y, target_idx)
    return X[mask], y[mask], lbl[mask]
```

```python
X_train_f, y_train_f, lbl_train_f = filter_split(X_train, y_train_idx, lbl_train, target_idx)
X_val_f, y_val_f, lbl_val_f = filter_split(X_val, y_val_idx, lbl_val, target_idx)
X_test_f, y_test_f, lbl_test_f = filter_split(X_test, y_test_idx, lbl_test, target_idx)
```

```python
X_train_mlp = X_train_f.reshape(X_train_f.shape[0], -1)
X_val_mlp   = X_val_f.reshape(X_val_f.shape[0], -1)
X_test_mlp  = X_test_f.reshape(X_test_f.shape[0], -1)
```

```python
print(X_train_mlp.shape)
print(X_val_mlp.shape)
print(X_test_mlp.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_mlp = scaler.fit_transform(X_train_mlp)
X_val_mlp   = scaler.transform(X_val_mlp)
X_test_mlp  = scaler.transform(X_test_mlp)
```

```python
y_train_mlp = y_train_f
y_val_mlp = y_val_f
```

```python
y_train_f.shape
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train_mlp.shape[1],)),

    tf.keras.layers.Dense(1024, activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Dense(256, activation='relu'),

    tf.keras.layers.Dense(len(np.unique(y_train_f)), activation='softmax')
])

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

```

```python
model.fit(
    X_train_mlp, y_train_mlp,
    validation_data=(X_val_mlp, y_val_mlp),
    epochs=20,
    batch_size=256
)
```

EXTRA:Additionally, in order to use them with your model the data should be randomly shuffled. An example of this procedure:

```python
from sklearn.utils import shuffle

np.random.seed(1961)

X_train, Y_train, lbl_train = shuffle(
    X_train[:], Y_train[:], lbl_train[:],
    random_state=1961
)

X_val, Y_val, lbl_val = shuffle(
    X_val[:], Y_val[:], lbl_val[:],
    random_state=1961
)

X_test, Y_test, lbl_test = shuffle(
    X_test[:], Y_test[:], lbl_test[:],
    random_state=1961
)
```
